```python
#Ejercicio2: calcular ventas por region
print("Ejercicio2: calcular ventas por region")
Ventas_por_region = df.groupby('Region')['Ventas'].sum()
print(Ventas_por_region)
print("¨/n")
print(df.head(10))
```

```python
# Ejercicio 3:filtrar el df por las ventas de david
print("Ejercicio3:filtrar el df por las ventas de david")
Ventas_david = df[df['Vendedor']=='David']
print(Ventas_david)
print("/n")
```

```python
# Ejercicio 4: Promedio de ventas por producto
print("Ejercicio 4: Promedio de ventas por producto")
Ventas_por_producto = df.groupby('Producto')['Ventas'].mean()
print(Ventas_por_producto)
print("/n")
```

```python
# Ejercicio 5: orden df por ventas en orden decendente
print("Ejercicio 5: orden df por ventas en orden decendente")
df_por_orden = df.sort_values(by='Ventas', ascending=False)
print(df_por_orden)
```

```python
#Ejercicio 6 : filtrar filass co ventas mayores a 10.000
print('Ejercicio 6 : filtrar filass co ventas mayores a 10.000')
df_filtrado= df[df['Ventas']>10000]
print(df_filtrado)
print("/n")
```

```python
#Ejercicio 7: orde el df en orden desendente
print("Ejercicio 7: orden df por ventas en orden decendente")
df_por_orden = df.sort_values(by='Ventas', ascending=True)
print(df_por_orden)
```

```python
# Ejercicio 8: eliminar la columna vendedor
print('Ejercicio 8: eliminar la columna vendedor')
df_sin_vendedor = df.drop(columns=['Vendedor'])
print(df_sin_vendedor)
print("/n")
```
